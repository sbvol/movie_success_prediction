```python
df = df.loc[1:]
```

```python
x1 = df.iloc[1:13,0:1]
y1 = df.iloc[1:13,1:2]
y2 = df.iloc[1:13,2:3]
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10,8))
ax1.plot(x1,y1)
ax1.set_title('Tickets Sold over Year')
ax2.set_title('Box Office Dollars')
ax2.plot(x1,y2)
plt.xlabel('year')
plt.tight_layout()
plt.show()
```

```python
x1 = df2.iloc[2:13,0:1]
y1 = df2.iloc[2:13,1:9]
y2 = df2.iloc[2:13,7:8]
y3 = df2.iloc[2:13,8:9]
labels1 = ['Warner Bros', 'Walt Disney', '20th Century Fox',\
          'Paramount Pictures', 'Sony Pictures', 'Universal',\
          'Total Major 6', 'Total Other Studios']
labels2 = ['Total Major 6', 'Total Other Studios']
```

```python
labels2[0]
```

```python
fig,ax = plt.subplots(1,1, figsize=(10,8))
ax.plot(x1,y1)
ax.legend(loc='best')
ax.legend(labels1)
plt.xlabel('year')
plt.ylabel('number of movies')
plt.title('Number of movies distributed per year')
plt.show()
```

```python
fig,(ax1,ax2) = plt.subplots(2,1, figsize=(10,5), sharex=True)
ax1.plot(x1,y2)
ax2.plot(x1,y3)
ax1.legend(loc='best')
ax2.legend(loc='best')
ax1.legend(labels2[0])
ax2.legend(labels2[1])
plt.xlabel('year')
#plt.ylabel('number of movies')
#plt.title('Number of movies distributed per year')
plt.show()
```
